```python
import numpy as np

time = 1941
end = 1945

dt = 1/365.0;

N = int((end - time) / dt)

ng = [0]
ns = [0]

ng[0] = 8500000 # count of Germany soldiers
ns[0] = 5100000 # count of USSR soldiers

alfaG = 0.35 
alfaS = 0.40

betaG = 0.1
betaS = 0.6

gammaG = 35000
gammaS = 1000

i = 0

while time <= end:

    ng.insert(i + 1, (dt * (-alfaG * ng[i] - betaS * ns[i] + gammaG) + ng[i]))
    ns.insert(i + 1, (dt * (-alfaS * ns[i] - betaG * ng[i] + gammaS) + ns[i]))
    
    if ng[i + 1] < 0 :
        ng[i + 1] = 0.0
        print("USSR winner with",int(ns[i + 1]),"soldiers, time is", round(time,2))
        break
    elif ns[i + 1] < 0 :
        ns[i + 1] = 0.0
        print("GERMANY winner",int(ng[i + 1]),"soldiers, time is", round(time,2))
        break
    elif i + 1 == 1461:
        print("Nobody won")
    i += 1;
    time = time + dt

import matplotlib.pyplot as plt
x = np.linspace(1941, end, N)

plt.plot(ng, label = "Germany")

plt.plot(ns, label = "USSR")

plt.legend()
plt.show()
```
